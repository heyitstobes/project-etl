```python
import pandas as pd
```

```python
uninsured_df = pd.read_excel('clean_data/insurance_coverage_by_state_by_year.xlsx')
```

```python
uninsured_df.head()
```

```python
uninsured_df = uninsured_df.loc[uninsured_df['Coverage'] == 'Uninsured',:]
uninsured_df = uninsured_df.loc[uninsured_df['Year'] <= 2010,:].sort_values('Year').reset_index()
uninsured_df = uninsured_df[['Year','Nation/State','Percent']]
uninsured_df = uninsured_df.rename(columns={'Year':'poll_year','Nation/State':'location','Percent':'rate_of_uninsured'})
```

```python
uninsured_df.to_csv('clean_data/uninsured.csv')
```
